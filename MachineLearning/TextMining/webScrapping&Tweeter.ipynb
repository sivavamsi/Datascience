```python
res_tag = browser.find_elements_by_css_selector('div.h100.pb20')
len(res_tag)
```

```python
restaurants = pd.DataFrame()
for res in res_tag:
    title = res.find_element_by_css_selector('div.res_title')
    title = title.get_attribute('title')
    
    loc = res.find_element_by_css_selector('div.nowrap.grey-text.fontsize5.ttupper')
    loc = loc.text
    
    food_type = res.find_element_by_css_selector('div.ptop0.pbot0.pl10.pr10')
    food_type = food_type.find_elements_by_css_selector('div')[2].text
    
    try:
        url = res.find_element_by_css_selector('a[data-link-type="restaurant"]')
        url = url.get_attribute('href')
    except:
        url = None
    
    curr_res = {'title':title,'loc':loc,'food_type':food_type,'url':url}
    restaurants = restaurants.append(curr_res,ignore_index=True)
restaurants.shape
    #food_type = res.find_element_by_css_selector('div.po')
    #or we can use title.text
    #print(title,loc,food_type,url)
    #print('------')
```
