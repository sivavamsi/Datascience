```python
from selenium import webdriver
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from tqdm import trange, tqdm
from time import sleep
import time
from selenium.webdriver.chrome.options import Options
print("webdriver version: ",webdriver.__version__)
```

```python
# browser = webdriver.Chrome('chromedriver.exe')
```

```python

browser = webdriver.Chrome(ChromeDriverManager().install())
    
browser.get('http://google.com')
time.sleep(5)
```

```python
def scroll_down(driver):
    start = time.time()
    SCROLL_PAUSE_TIME = 10
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(SCROLL_PAUSE_TIME)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    print("{} seconds to take full scrolling web the page".format(round(time.time()-start,3)))
```

```python
url = "https://www.zomato.com/bangalore/order-food-online"
browser.get(url)
time.sleep(10)
scroll_down(browser)
time.sleep(15)
```

```python
dict_df = []
div_tag = browser.find_elements_by_css_selector('div.jumbo-tracker')
for i in tqdm(range(len(div_tag))):
    try:
        resturant_name = div_tag[i].find_elements_by_tag_name('h4')[0].text
        a_block = div_tag[i].find_elements_by_tag_name('a')
        p = a_block[0].find_elements_by_tag_name('p')
        try:
            offer = a_block[0].find_elements_by_tag_name('div')[-1].text
        except:
            offer = ''
        deliverytime = p[-1].text
        if offer == deliverytime:
            deliverytime=''
        p = a_block[1].find_elements_by_tag_name('p')
        Speciality = p[0].text
        Price = p[1].text
    
        temp = {"Resturant":resturant_name,
               "Offer":offer,
               "Delivery Time":deliverytime,
               "Speciality":Speciality,
               "Price":Price,
               }
        dict_df.append(temp)
    except:
        pass
browser.close()
```

```python

df = pd.DataFrame(dict_df)
df.info()
```
